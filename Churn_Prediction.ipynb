# **II. Data Cleaning**


## 1.Transformation of some feature's type


```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
print(data.dtypes)
```

### 2. Detection of missing values


```python
data.isnull().sum()
```

```python
>> As we see here, our dataset contains 11 missing values on 'TotalCharges'
```

## 3.Handling missing values


>> We will replace the missing values with the median of the features

```python
median_total_charges = data['TotalCharges'].median()
data['TotalCharges'].fillna(median_total_charges, inplace=True)

print(data.isnull().sum())
```




# III. Transformation of the dataset





 ## 1. Transforming categorical attributes into numeric attributes
 

```python
from sklearn.preprocessing import LabelEncoder
for c in data.columns:
    if data[c].dtype == 'object':
        labelencoder = LabelEncoder()
        data[c] = labelencoder.fit_transform(data[c])
```

```python
data.dtypes
```

```python
data.head(10)
```

>>> all of values are numeric

## 2.Seperating data



```python
x=data.drop(['Churn'],axis=1)
y=data['Churn']
```

## 3.Normalizing values


```python
from sklearn import preprocessing
x_normalize=preprocessing.StandardScaler()
x_norm=x_normalize.fit_transform(x)
```

```python
y_array = y.to_numpy()
#normalizing of y
y_normalize=preprocessing.MinMaxScaler()
y_norm=y_normalize.fit_transform(y_array.reshape(-1, 1))
```

```python
x_norm,y_norm
```

```python
>> All values are normalized
```

# **IV. Creating and validating a model**


```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x_norm,y,test_size=0.1,random_state=42,stratify=y)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Utiliser un modèle de régression logistique
model = LogisticRegression(random_state=42)
model.fit(x_train, y_train)

# Prédire sur l'ensemble de test
y_predict = model.predict(x_test)

# Calculer le score d'exactitude
accuracy = accuracy_score(y_test, y_predict)
print(f"Accuracy Score: {accuracy}")
```
